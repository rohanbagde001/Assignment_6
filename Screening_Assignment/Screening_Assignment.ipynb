# Multiple Inheritance

```python
class A:
    def method1(self):
        print("Class A inside method1")
```

```python
## Class B is inheriting base class A

class B(A):
    def method1(self):
        print("class B is inside method1")
        

## class C inheriting base class A

class C(A):
    def method1(self):
        print("class C is inside class method1")
    def method2(self):
        print("class C is inside class method2")
```

```python
## class D inheritting base class B and C

class D(B,C):
    def method1(self):
        print('class D is inside method1')
```

```python
obj_d = D()
```

```python
obj_d.method1()
obj_d.method2()
```

```python
B.method1(obj_d)
```

```python
A.method1(obj_d)
```

#  Decorator

```python
Decorator allows to modify the behaviour of a function or a class.

Decorator allow us to wrap another function in order to extend the behaviour of the wrapped
function, without permanently modify it
```

```python
def decorator(func):
    def inner_function(*args,**kwargs):
        
        print("Before Execution")
        
        returned_value = func(*args, **kwargs)
        
        print("After Excecution")
        
        return returned_value
        
    return inner_function


# adding decorator to the function
@decorator
def sum_two_numbers(a, b):
    print("Inside the function")
    return a + b

# getting the value through return of the function
print("Sum =", sum_two_numbers(a=1, b=2))
```
