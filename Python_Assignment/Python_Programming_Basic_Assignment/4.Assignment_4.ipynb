```python
1. Write a Python Program to Find the Factorial of a Number?

Ans:-
```

```python
def factorial(num):
    if (num < 1):
        return 1
    else:
        return num*factorial(num-1)
num  = int(input("Enter a number: ")) 
value = factorial(num)
print(f'The factorial of {num} is {value}')
```

```python
2.Write a Python Program to display the multiplication table ?

Ans:-
```

```python
def generateTable(base,entries):
    for i in range(1,entries+1):
        print(f'{base} X {i} = {base*i}')
        
num = int(input("Enter a number: "))
values = int(input("Enter no of entries: "))
generateTable(num,values)
```

```python
s_count = int(input('Enter the number of fibonacci sequence you want? '))
initial_list = [0,1]
if s_count < 0:
    print("Fibonacci Numbers are not available for Negetive Numbers")
elif s_count <=2 and s_count>=0:
    print(initial_list)
else:
    for i in range(s_count):
        if i>=2:
            initial_list.append(initial_list[i-1]+initial_list[i-2])
            print(f' The first {s_count} fibonacci series are: ',initial_list)
```

```python
4.Write a Python Program to check Armstrong number ?

Ans:-
    
```

```python
def checkArmstrongNumber():
    in_num = input('Enter a number:')
    sum = 0
    for char in range(len(in_num)):
        sum = sum+ pow(int(in_num[char]),3)
    if sum == int(in_num):
        print(f'{in_num} is an Armstrong Number')
    else:
        print(f'{in_num} is not a Armstrong number')
for i in range(2):
    checkArmstrongNumber()
```

```python
5.Write a Python Program to Find Armstrong number in an interval ?

Ans:-
    
```

```python
def checArmstrongNumber(in_num, storage):
    sum = 0
    for char in range(len(in_num)):
        sum = sum + pow(int(in_num[char]),3)
    if sum ==int(in_num):
        storage.append(int(in_num))
        
start_interval = int(input("Enter the start of the interval: "))
end_interval = int(input("Enter the end of the interval: "))
list_of_armstrong = []

if start_interval > end_interval:
    print("Start interval cannot be greater than the End Interval")
else:
    for number in range(start_interval,end_interval+1):
        checArmstrongNumber(str(number),list_of_armstrong)
        print(f'The Armstrong numbers between {start_interval} and {end_interval} are {list_of_armstrong}')
        
```

```python
def sumofNaturalNumbers(num):
    sum = num*((num+1)/2)
    print(f'sum of {num} natural numbers is {sum}')
        
num = int(input("Enter a Number: "))
sumofNaturalNumbers(num)
```
